```python
from fuzzywuzzy import process
```

```python
product_list = [
    "Mysore Traditional Paintings", "Madhubani Paintings", "Thanjavur Paintings",
    "Nirmal Paintings", "Cheriyal Paintings", "Warli Painting", "Kangra Paintings",
    "Sohrai â€“ Khovar Painting", "Karuppur Kalamkari Paintings", "Gond Painting of Madhya Pradesh",
    "Basohli Painting", "Nathdwara Pichhwai Painting", "Painting of Lanjia Saura (IDITAL), Odisha",
    "Assam Majuli Manuscript Painting", "Banaras Mural Painting"
]
```

```python
def match_products(description, product_list, threshold=80):
    match, score = process.extractOne(description, product_list)
    if score >= threshold:
        return match
    else:
        return None

# Apply the function to the Product Description column
df['Matched Product'] = df['Product Description'].apply(match_products, args=(product_list,))

# Filter out rows where no match was found
matched_df = df.dropna(subset=['Matched Product'])

# Calculate the market potential per unit for each matched product
matched_df["Market Potential per Unit"] = matched_df["Estimated F.O.B Value $"] / matched_df["Standard Qty"]

# Aggregate to find the total market potential for each matched product
market_potential = matched_df.groupby("Matched Product").agg({
    "Estimated F.O.B Value $": "sum",
    "Standard Qty": "sum",
    "Market Potential per Unit": "mean"
}).reset_index()

# Sort by total Estimated F.O.B Value to find the product with the highest market potential
market_potential = market_potential.sort_values(by="Estimated F.O.B Value $", ascending=False)

print("\nMarket Potential for Each Product:")
print(market_potential)

# Find the product with the highest market potential
top_product = market_potential.iloc[0]

print("\nProduct with Highest Market Potential:")
print(top_product)
```

Steps for Competitor Analysis
1.Identify Key Competitors: Determine the major shippers of Warli paintings.
2.Market Share Analysis: Calculate the market share of each shipper based on quantity and value.
3.Export Patterns: Analyze the export patterns of these shippers, including destinations and shipment frequency.
4.Pricing Strategies: Compare the unit rates and total values of shipments to understand pricing strategies.
5.Growth and Trends: Evaluate trends over time to see if any shipper is gaining or losing market share.

```python
# Calculate total quantity and value for each shipper
shipper_summary = warli_df.groupby('Shipper Name').agg({
    'Standard Qty': 'sum',
    'Estimated F.O.B Value $': 'sum'
}).reset_index()

# Calculate market share
total_qty = shipper_summary['Standard Qty'].sum()
total_value = shipper_summary['Estimated F.O.B Value $'].sum()

shipper_summary['Market Share Qty'] = (shipper_summary['Standard Qty'] / total_qty) * 100
shipper_summary['Market Share Value'] = (shipper_summary['Estimated F.O.B Value $'] / total_value) * 100

print(shipper_summary)
```

```python
# Export patterns - shipments per month
shipments_per_month = warli_df.groupby(['Shipper Name', 'Month']).size().unstack(fill_value=0)

print(shipments_per_month)
```

```python
# Average unit rate per shipper
average_unit_rate = warli_df.groupby('Shipper Name')['Standard Unit Rate $'].mean().reset_index()

print(average_unit_rate)
```

```python
# Trends over time - total value per month
monthly_trends = warli_df.groupby(['Shipper Name', 'Month'])['Estimated F.O.B Value $'].sum().unstack(fill_value=0)

print(monthly_trends)
```

```python
# Calculate total quantity and value for each shipper
shipper_summary = warli_df.groupby('Shipper Name').agg({
    'Standard Qty': 'sum',
    'Estimated F.O.B Value $': 'sum'
}).reset_index()

# Get the top 5 shippers based on quantity
top_5_shippers = shipper_summary.nlargest(5, 'Standard Qty')

# Filter the original DataFrame for only the top 5 shippers
top_5_warli_df = warli_df[warli_df['Shipper Name'].isin(top_5_shippers['Shipper Name'])]

# Update shipper summary for the top 5 shippers
top_5_shipper_summary = top_5_warli_df.groupby('Shipper Name').agg({
    'Standard Qty': 'sum',
    'Estimated F.O.B Value $': 'sum'
}).reset_index()

# Calculate market share
total_qty = top_5_shipper_summary['Standard Qty'].sum()
total_value = top_5_shipper_summary['Estimated F.O.B Value $'].sum()

top_5_shipper_summary['Market Share Qty'] = (top_5_shipper_summary['Standard Qty'] / total_qty) * 100
top_5_shipper_summary['Market Share Value'] = (top_5_shipper_summary['Estimated F.O.B Value $'] / total_value) * 100

print(top_5_shipper_summary)
```

```python
# Visualization of market share by quantity
plt.figure(figsize=(10, 6))
sns.barplot(data=top_5_shipper_summary, x='Shipper Name', y='Market Share Qty')
plt.title('Market Share by Quantity for Top 5 Shippers of Warli Paintings')
plt.xlabel('Shipper Name')
plt.ylabel('Market Share (%)')
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
# Visualization of market share by value
plt.figure(figsize=(10, 6))
sns.barplot(data=top_5_shipper_summary, x='Shipper Name', y='Market Share Value')
plt.title('Market Share by Value for Top 5 Shippers of Warli Paintings')
plt.xlabel('Shipper Name')
plt.ylabel('Market Share (%)')
plt.xticks(rotation=45, ha='right')
plt.show()
```
